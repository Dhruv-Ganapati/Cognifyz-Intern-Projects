```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('Dataset .csv')
```

**missing values and encoding categorical variables.**

**-Missing Values**

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Out of 9551 values, 9 values are missing. So we need to fill those values by most repeating values
# Using mode()

# Filling "North Indian" in  Null position
fill_cuisines = df['Cuisines'].mode()[0]
df['Cuisines'] = df['Cuisines'].fillna(fill_cuisines)
```

```python
# Drop ID columns - Useless
df.drop(['Restaurant ID'], axis=1, inplace=True)
```

```python
# Checking if there exist any null values
# Filling "North Indian" in  Null position
df.isnull().sum()
```

```python
df.shape
```

**These are the following unique values in each columns.**

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

* Here, there are 1825 different cuisines.

```python
df.astype('object')
```

**Removing single instances class**

```python
c_counts = df['Cuisines'].value_counts()
single_instance_class = c_counts[c_counts == 1].index
df = df[~df['Cuisines'].isin(single_instance_class)]
```

```python
# df.drop('Restaurant ID', inplace=True)
df.columns.to_list()
```

```python
df.shape
```

```python
# Feature columns and target columns
feature_cols = ['Country Code', 'City', 'Locality', 'Locality Verbose', 'Longitude',
                'Latitude', 'Average Cost for two', 'Currency',
                   'Has Table booking', 'Has Online delivery', 'Is delivering now',
                   'Switch to order menu', 'Price range', 'Aggregate rating',
                   'Rating color', 'Rating text', 'Votes']

target_col = 'Cuisines'
```

```python
X= df[feature_cols]
y= df[target_col]
```

```python
# Encoding categorical features and target variable
# With Numerical- StandardScaler and Categorical- ONeHotEncoder distribution

preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), ['Longitude', 'Latitude', 'Price range',
                               'Average Cost for two', 'Aggregate rating', 'Votes']),
    ('cat', OneHotEncoder(handle_unknown='ignore'), ['Country Code', 'City', 'Locality', 'Locality Verbose', 'Currency',
                                  'Has Table booking', 'Has Online delivery', 'Is delivering now',
                                  'Switch to order menu', 'Rating color', 'Rating text'])
])
```

*There is a chance of presence of new or unseen categories in test set that were not present in training set. **So we use handle_unknown='ignore' parameter**.*

```python
# Encode the target variables
label_encoder = LabelEncoder()
y_encoder = label_encoder.fit_transform(y)
```

```python
# Splitting datas into Training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoder, test_size=0.2, random_state=60)
```

**Random Forest Classification Algorithm**

```python
# Random Forest Classifiers Model defining
model = Pipeline(steps=[
    ('preprocessor', preprocessor), ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
model.fit(X_train, y_train)
```

```python
# Make prediction on test data
y_pred = model.predict(X_test)
```

**Evaluating Model Performance:**
* Accuracy
* Precision
* Recall

```python
print(classification_report(y_test, y_pred, labels=np.arange(len(label_encoder.classes_)), target_names=label_encoder.classes_))
```

**accuracy,
precision, recall)**

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred, labels=np.arange(len(label_encoder.classes_)))
```

```python
conf_matrix
```

```python
report = classification_report(y_test, y_pred, labels=np.arange(len(label_encoder.classes_)), target_names=label_encoder.classes_, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)
```

```python
# cuisines with significantly lower performance metrics
low_performance_cuisines = report_df[(report_df['precision'] < 0.5) | (report_df['recall'] < 0.5)]
print("Cuisines with potential challenges or biases: ", low_performance_cuisines)
```
